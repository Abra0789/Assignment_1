```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as pt
import seaborn as sns
from sklearn.model_selection import  train_test_split as tts
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error, classification_report
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_excel("cancer.xlsx")
```

```python
df.head()
```

```python
df.drop('id', axis=1, inplace=True)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df['diagnosis'] = le.fit_transform(df['diagnosis'])
```

```python
df.head()
```

```python
x = df.drop('diagnosis', axis=1)
```

```python
x.head()
```

```python
y = df[['diagnosis']]
```

```python
x.head()
```

```python
xtrain, xtest, ytrain, ytest = tts(x,y, test_size=0.3, random_state=50)
```

# Decision Tree

```python
clf = DecisionTreeClassifier()
```

```python
clf.fit(xtrain, ytrain)
```

```python
pred_clf = clf.predict(xtest)
```

```python
print(classification_report(ytest, pred_clf))
```

```python
mse = mean_squared_error(ytest, pred_clf)
```

```python
mae = mean_absolute_error(ytest, pred_clf)
```
